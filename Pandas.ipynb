```python
# to delete a column
df = df.drop(columns=['new score'])
df
```

```python
# add row to our dataset
df.loc[len(df.index)] = ['female', 'group D', 'some college','free/reduced','none',77,86,86]
df
```

```python
# check duplicate values it will return true if there is a duplicate values
print(df.duplicated())
```

```python
# remove duplicate values
df.drop_duplicates(inplace = True)
df
```

```python
# to check none value in test preparation course
df[df['test preparation course']== 'none']
```

```python
# change the none value in test preparation course to another value (cleaning data)
for x in df.index:
  if df.loc[x, "test preparation course"] == 'none':
    df.loc[x, "test preparation course"] = 'progress'
df
```

```python
# delete rows of the none values in test preparation course
for x in df.index:
  if df.loc[x, "test preparation course"] == 'progress':
    df.drop(x, inplace = True)
df['test preparation course']
```

```python
# delete the column
df = df.drop(columns='test preparation course')
df
```

```python
# change the gender into numerical value male=0 & female=1
df['gender'].replace(['male', 'female'],
                        [0, 1], inplace=True)

df
```

```python
# to convert catigorical datasets into numerical values
dummies = pd.get_dummies(df['race/ethnicity'])

# Concatenate the dummies to original dataframe
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
# to find relationship between our datasets we use corr()
merged.corr()
```
