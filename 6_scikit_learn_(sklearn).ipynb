# Standard Scaler

```python
# import libraries
import numpy as np
from sklearn.preprocessing import StandardScaler

# define matrix with two column the first is x and the second is y
data = np.array([[345, 100],
                [160, 50],
                [330, 40],
                [60, 76543]])

# then we will fit the standardscaler
scaler = StandardScaler().fit(data)
```

```python
# we will see the mean
scaler.mean_
```

```python
# we will see the variance
scaler.scale_
```

```python
# here we will transform the data into the same range
scaled_data = scaler.transform(data)

# show data before and after scale
print('data : \n',data)
print("________________________________________________")
print('\n scaled data : \n',scaled_data)
print("________________________________________________")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1,2)
axes[0].set_title('without standardscatter')
sns.scatterplot(data = data, x = data[:,0], y = data[:,1],ax=axes[0])
axes[1].set_title('with standardscatter')
sns.scatterplot(data = scaled_data, x = scaled_data[:,0], y = scaled_data[:,1],ax=axes[1])
plt.show()
```

# Robust Scaler

```python
# import libraries
import numpy as np
from sklearn.preprocessing import RobustScaler

# define matrix with two column the first is x and the second is y
data = np.array([[345, 100],
                [160, 50],
                [330, 40],
                [60, 76543]])

# then we will fit the RobustScaler
scaler = RobustScaler().fit(data)
```

# Label Encoding

```python
import numpy as np
import pandas as pd

# define data
data =np.array([['good','excellent','a'],['bad','good','b'],
                ['good','bad','c'],['good','good','b']])

# transform data to pandas dataframe
df = pd.DataFrame(data)
df.head()
```

```python
# give the columns names
df.columns = ['math','science','class']
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# apply encoding on label
label_encoder = LabelEncoder()
df['class'] = label_encoder.fit_transform(df['class'])

df.head()
```

# Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder

# apply encoding on feature
oe = OrdinalEncoder(categories= [['bad','good','excellent'],['bad','good','excellent']])
df[['math','science']] = oe.fit_transform(df[['math','science']])

df.head()
```

```python
# import libraries
import pandas as pd
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()

# Convert data to pandas dataframe
diabetes_df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)

# Add the target label
diabetes_df["target"] = diabetes.target

# show the head
diabetes_df.head()
```

```python
diabetes_df.tail()
```

```python
diabetes_df.shape
```

```python
diabetes_df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig, axes = plt.subplots(3, 3, figsize=(16, 16))

# Use the axes for plotting
sns.scatterplot(x = diabetes_df['age'], y = diabetes_df['target'], ax = axes[0,0])
sns.scatterplot(x = diabetes_df['sex'], y = diabetes_df['target'], ax = axes[0,1])
sns.scatterplot(x = diabetes_df['bmi'], y = diabetes_df['target'], ax = axes[0,2])

sns.scatterplot(x = diabetes_df['s1'], y = diabetes_df['target'], ax = axes[1,0])
sns.scatterplot(x = diabetes_df['s2'], y = diabetes_df['target'], ax = axes[1,1])
sns.scatterplot(x = diabetes_df['s3'], y = diabetes_df['target'], ax = axes[1,2])

sns.scatterplot(x = diabetes_df['s4'], y = diabetes_df['target'], ax = axes[2,0])
sns.scatterplot(x = diabetes_df['s5'], y = diabetes_df['target'], ax = axes[2,1])
sns.scatterplot(x = diabetes_df['s6'], y = diabetes_df['target'], ax = axes[2,2])
```

```python
X = diabetes_df[diabetes.feature_names].copy()
y = diabetes_df["target"].copy()
```

```python
from sklearn.model_selection import train_test_split

# Split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.15,
                                                                   random_state=999)
```

```python
from sklearn.linear_model import LinearRegression

# train the model on training set
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# make prediction with test data
y_test_pred = linear_model.predict(X_test)
```

```python
import sklearn.metrics as sm
print("Mean absolute error =", round(sm.mean_absolute_error(y_test, y_test_pred), 2))
print("Mean squared error =", round(sm.mean_squared_error(y_test, y_test_pred), 2))
print("Median absolute error =", round(sm.median_absolute_error(y_test, y_test_pred), 2))
print("Explain variance score =", round(sm.explained_variance_score(y_test, y_test_pred), 2))
print("R2 score =", round(sm.r2_score(y_test, y_test_pred), 2))
```

```python
coeff = pd.Series(linear_model.coef_, index = X_train.columns)
intercept = linear_model.intercept_
```

```python
print("Coefficients:\n")
print(coeff)
print("\n")
print("Intercept:\n")
print(intercept)
print("\n")
```
