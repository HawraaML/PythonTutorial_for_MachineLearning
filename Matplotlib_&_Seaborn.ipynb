# Linear Charts

```python
''' linear charts with matplotlib
In this cell we will represent the relation between size of house and the price in linearchart

 1- we will import numpy library to make 1D array for both size and price
 2- then we will use matplotlib to visualize them'''

import numpy as np
house_area = np.array([100,150,200,250,300,350,400,450,500])
house_price = np.array([1000,2000,2500,4000,3000,5000,5400,4200,9000])

plt.plot(house_area, house_price)
plt.show()
```

```python
import numpy as np
house_area = np.array([100,150,200,250,300,350,400,450,500])
house_price = np.array([1000,2000,2500,4000,3000,5000,5400,4200,9000])

plt.plot(house_area, house_price)

# to add label to the chart
plt.xlabel('area')
plt.ylabel('price')
plt.title('Houses')
plt.show()
```

```python
'''Here we will draw two chart in the same figure
1- the first one is the degree of math to the average
2- the second one is the degree of science to the average'''

import numpy as np
math = np.array([55,60,77,80,90,99])
science = np.array([100,90,80,70,60,50])
average = np.array([60,75,80,85,90,99,])

plt.plot(math, average, marker= 'o')
plt.plot(science, average, marker= 's')
# to add label to the chart
plt.xlabel('degree of math and science')
plt.ylabel('average')
plt.title('Exam')

# legend use to add name for both line chart to distinguish between them
plt.legend(['math','science'])
plt.show()
```

```python
'''fmt is a convenient way for defining basic formatting like
color, marker and linestyle'''

import numpy as np
math = np.array([55,60,77,80,90,99])
science = np.array([100,90,80,70,60,50])
average = np.array([60,75,80,85,90,99,])

#the fmt format 'marker linestyle color'
plt.plot(math, average, 'o-g')
plt.plot(science, average, 's:c')

plt.xlabel('degree of math and science')
plt.ylabel('average')
plt.title('Exam')
plt.legend(['math','science'])
plt.show()
```

```python
# grid()

import numpy as np
math = np.array([55,60,77,80,90,99])
science = np.array([100,90,80,70,60,50])
average = np.array([60,75,80,85,90,99,])

plt.plot(math, average, 'o-g')
plt.plot(science, average, 's:c')

plt.xlabel('degree of math and science')
plt.ylabel('average')
plt.title('Exam')
plt.legend(['math','science'])
# we use grid to show them in the background of chart
plt.grid()
plt.show()
```

# Scatter plot

```python
#here we will plot the relation between number of student that took same degree in math
student_numm = np.array([5,7,8,7,2,17,20,9,19,11,12,9,6])
math_degree = np.array([99,86,87,88,100,86,103,87,94,78,77,85,86])
plt.scatter(student_numm, math_degree, color = 'red')


student_nums = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
science_degree = np.array([100,99,84,90,90,99,95,96,94,100,79,60,91,80,85])
plt.scatter(student_nums, science_degree, color = 'green')


plt.xlabel('number of student')
plt.ylabel('math & science')
plt.title('Students and Degree')
plt.legend(['math','science'])
plt.show()
```

# Bar charts

```python
# here we use bar char to show the relation between number of students and the class
classes = np.array(["A", "B", "C", "D"])
students = np.array([15, 20, 10, 25])

plt.bar(classes,students)
plt.show()
```

```python
# horizontal bar chart
classes = np.array(["A", "B", "C", "D"])
students = np.array([15, 20, 10, 25])

plt.barh(classes, students , color = 'green')
plt.xlabel('students')
plt.ylabel('classes')
plt.title('students in classes')
plt.show()
```

```python
# bar chart with seaborn
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.barplot(data = titanic, x = 'alone', y = 'fare', hue='pclass')
```

# Heatmaps

```python
flights = sns.load_dataset("flights").pivot("year", "month", "passengers")
flights.head()
```

```python
green = sns.light_palette("seagreen", as_cmap=True)
sns.heatmap(flights, fmt="d", annot=True, cmap = green)
```
